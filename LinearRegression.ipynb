```python
%matplotlib inline
import matplotlib.pyplot as plt

x1 = [0,1,2,3,4]
y1 = [0,0.5,2,3.5,4.5]

plt.plot(x1,y1,linestyle="",marker="o",color='r')
plt.show()
```

```python
def gety(m,c,x):
    return m*x+c

def draw_line(m,c):
    x = []
    y = []
    for point in range(5):
        x.append(point)
        y.append(gety(m,c,point))
    plt.plot(x,y,marker="o")    
    return (x,y)

(x,y) = draw_line(0.5,3)
```

```python
def get_error(m,c,x,y):    
    from sklearn.metrics import mean_squared_error
    sum = 0
    x1 = []
    y1 = []
    for i in range(len(y)):
        #squared_error = (y[i] - y1[i])*(y[i] - y1[i])
        squared_error = (gety(m,c,x[i]) - y[i])*(gety(m,c,x[i]) - y[i])
        sum += squared_error
        x1.append(x[i])
        y1.append((gety(m,c,x[i])))
    print 'sum of squared error', sum
    print 'mean sum of squared error', sum/len(y)
    print 'python mean sum of squared error', mean_squared_error(y1, y)
    plt.plot(x,y,linestyle="",marker="o",color='r')
    plt.plot(x1,y1,marker="o")
    return sum/len(y)

error = get_error(0.5,5,x1,y1)
print 'mean sum of squared error', error
```

```python
import random    

def optimize_line(x, y):
    for i in range(100):
        m = random.randint(-10, 10)
        print m
        c = random.randint(-10, 10)
        print c
        best_error = 1000000
        best_m = 1000000
        best_c = 1000000
        error = get_error(m, c, x, y)
        if error < best_error:
            best_error = error
            best_m = m
            best_c = c
    return(best_m, best_c, best_error)

(m, c, error) = optimize_line(x1, y1)
print 'best random line has m =', m, ' c =', c, 'and error =',error
```
