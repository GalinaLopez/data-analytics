```python
# gradient descent

# From calculation, we expect that the local minimum occurs at x=1

x_old = -1 # The value does not matter as long as abs(x_new - x_old) > precision
x_new = 0 # The algorithm starts at x=0
gamma = 0.1 # step size
precision = 0.000000

def f_derivative(x):
    return 2 * x - 2

def f(x):
    return x**2 - 2*x + 1

gradient_x = []
gradient_y = []
plot(x_new, f(x_new), marker = 'o', color = 'b')
gradient_x.append(x_new)
gradient_y.append(f(x_new))
while abs(x_new - x_old) > precision:
#while abs(f(x_new) - f(x_old)) > precision:
    x_old = x_new   
    x_new = x_old - gamma * f_derivative(x_old)
    plot(x_new, f(x_new), marker = 'o', color = 'b')
    gradient_x.append(x_new)
    gradient_y.append(f(x_new))
    print x_new
    print 'der',f_derivative(x_old)
    print x_old
    print 'abs',abs(x_new-x_old)
    
plot(x_new, f(x_new), marker = 'o', color = 'r')    
print("Local minimum occurs at", x_new)
```
